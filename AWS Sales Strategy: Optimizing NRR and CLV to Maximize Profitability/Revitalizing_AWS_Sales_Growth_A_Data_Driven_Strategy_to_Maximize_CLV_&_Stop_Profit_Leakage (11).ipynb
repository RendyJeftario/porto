```python
# --- 1. Setup Data ---
# Cost & List Price
df['Cost'] = df['Sales'] - df['Profit']
df['List_Sales'] = df.apply(lambda row: row['Sales'] / (1 - row['Discount']) if row['Discount'] < 1 else 0, axis=1)

# Identify "At Risk" Customers (Those with discount > 20%)
# These are the ones we are forcing to pay more. They might leave.
risk_mask = df['Discount'] > 0.20
df_risk = df[risk_mask].copy()
df_safe = df[~risk_mask].copy()

# --- 2. Calculate Scenarios ---

# Base Profit (Safe customers) - unaffected by policy
safe_profit = df_safe['Profit'].sum()

# Scenario B Profit for Risk Group (If 100% accept the new price)
# Apply Max Disc 20% rule
df_risk['New_Discount'] = 0.20
df_risk['New_Sales'] = df_risk['List_Sales'] * (1 - df_risk['New_Discount'])
df_risk['New_Profit'] = df_risk['New_Sales'] - df_risk['Cost']
potential_risk_profit = df_risk['New_Profit'].sum()

# Total Profit Scenario B
profit_scenario_b = safe_profit + potential_risk_profit


profit_scenario_a = 442528 # Taken from previous execution

# --- 3. Churn Simulation ---
# What happens if X% of the "Repriced" customers walk away (Churn)?
churn_rates = np.linspace(0, 1, 11) # 0% to 100% churn
simulated_profits = []

for churn in churn_rates:
    # If churn% leave, we only keep (1-churn)% of the potential profit from the risk group
    # Assumption: Churn is random among the risk group
    retained_risk_profit = potential_risk_profit * (1 - churn)
    total_sim_profit = safe_profit + retained_risk_profit
    simulated_profits.append(total_sim_profit)

# --- 4. Visualization ---
plt.figure(figsize=(10, 6))
plt.plot(churn_rates * 100, simulated_profits, marker='o', linewidth=3, label='Skenario B dengan Churn')
plt.axhline(y=profit_scenario_a, color='orange', linestyle='--', linewidth=2, label='Skenario A (Stop Loss Only) - $442k')
plt.axhline(y=df['Profit'].sum(), color='red', linestyle='--', linewidth=2, label='Profit Saat Ini - $286k')

# Find Break-even Point (Where Scenario B drops below Scenario A)
# safe_profit + potential * (1 - churn) = profit_scenario_a
# potential * (1 - churn) = profit_scenario_a - safe_profit
# 1 - churn = (profit_scenario_a - safe_profit) / potential
# churn = 1 - (profit_scenario_a - safe_profit) / potential
break_even_churn = 1 - (profit_scenario_a - safe_profit) / potential_risk_profit

plt.title('Seberapa Banyak Customer Bisa "Kabur" Sebelum Skenario B Gagal?')
plt.xlabel('Persentase Customer "Mahal" yang Kabur (Churn Rate %)')
plt.ylabel('Total Profit ($)')
plt.legend()
plt.grid(True, alpha=0.3)

# Add text for break-even
if 0 <= break_even_churn <= 1:
    plt.axvline(x=break_even_churn*100, color='gray', linestyle=':')
    plt.text(break_even_churn*100 + 1, profit_scenario_a + 5000, f'Titik Impas: {break_even_churn*100:.1f}% Churn', rotation=0)

plt.savefig('churn_risk_analysis.png')

print(f"--- Analisis Resiko Skenario B ---")
print(f"1. Profit Ideal Skenario cut loss: ${profit_scenario_b:,.0f}")
print(f"2. Profit Aman diskon max 20%: ${profit_scenario_a:,.0f}")
print(f"3. Buffer Resiko: Anda bisa kehilangan hingga {break_even_churn*100:.1f}% dari customer yang dire-negosiasi")
print("   dan profitnya MASIH SAMA dengan Skenario A.")
```

```python
df['Order Month'] = df['Order Date'].dt.to_period('M').dt.to_timestamp()
df['Cohort Month'] = df.groupby('Customer ID')['Order Month'].transform('min')

def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    return year, month
order_year, order_month = get_date_int(df, 'Order Month')
cohort_year, cohort_month = get_date_int(df, 'Cohort Month')

df['Cohort Index'] = (order_year - cohort_year) * 12 + (order_month - cohort_month) + 1

```

###CRR (Customer Retention Rate)
Persentase jumlah pelanggan yang berhasil dipertahankan oleh perusahaan selama periode waktu tertentu.

```python
# --- 2. FUNGSI VISUALISASI FULL 48 BULAN ---
def plot_full_heatmap(dataframe, segment_name):
    # Filter Data
    segment_df = dataframe[dataframe['Segment'] == segment_name]

    # Pivot Table
    cohort_data = segment_df.groupby(['Cohort Month', 'Cohort Index'])['Customer ID'].nunique().reset_index()
    cohort_counts = cohort_data.pivot(index='Cohort Month', columns='Cohort Index', values='Customer ID')

    # Hitung Retensi
    cohort_sizes = cohort_counts.iloc[:, 0]
    retention = cohort_counts.divide(cohort_sizes, axis=0)

    # --- PENGATURAN VISUALISASI FULL ---

    # Menghitung dimensi yang dibutuhkan
    n_rows = len(retention)
    n_cols = retention.shape[1] # Ini akan sekitar 48

    # Lebar gambar harus besar agar 48 kolom muat.
    # Kita pakai rasio: 0.5 inch per kolom -> 48 * 0.5 = 24 inch lebar
    fig_width = max(20, n_cols * 0.6)
    fig_height = max(8, n_rows * 0.4 + 2)

    plt.figure(figsize=(fig_width, fig_height))

    plt.title(f'Customer Retention Rate (CRR): {segment_name} (All Months)', fontsize=20, pad=20)

    # Format Label YYYY-MM
    y_labels = [pd.to_datetime(t).strftime('%Y-%m') for t in retention.index]

    sns.heatmap(retention,
                annot=True,
                fmt='.0%',
                cmap='YlGnBu',
                vmin=0.0,
                vmax=0.5,
                yticklabels=y_labels,
                annot_kws={"size": 8}, # Perkecil font angka agar tidak tabrakan
                cbar_kws={'label': 'Retention Rate'},
                linewidths=0.1,
                linecolor='white'
               )

    plt.ylabel('Cohort Month', fontsize=14)
    plt.xlabel('Cohort Index (Months)', fontsize=14)
    plt.yticks(rotation=0)

    plt.tight_layout()
    plt.savefig(f'retention_heatmap_{segment_name}_full.png')
    plt.show()

# --- EKSEKUSI ---
segments = ['SMB', 'Strategic', 'Enterprise']
for seg in segments:
    plot_full_heatmap(df, seg)
```

```python
# 3. Calculate Overall Retention Curve per Segment (Full Duration)
segments = ['SMB', 'Strategic', 'Enterprise']
results = {}

for segment in segments:
    # Filter by segment
    seg_df = df[df['Segment'] == segment]

    # Calculate Cohort Counts
    cohort_data = seg_df.groupby(['Cohort Month', 'Cohort Index'])['Customer ID'].nunique().reset_index()
    cohort_counts = cohort_data.pivot(index='Cohort Month', columns='Cohort Index', values='Customer ID')

    # Calculate Retention Rate
    cohort_sizes = cohort_counts.iloc[:, 0]
    retention = cohort_counts.divide(cohort_sizes, axis=0)

    # Calculate Average Retention per Cohort Index (Column-wise mean)
    avg_retention = retention.mean(axis=0) * 100
    results[segment] = avg_retention

# Combine into a DataFrame
retention_summary = pd.DataFrame(results)

# Fill NaNs with 0 for plotting (though logically retention stops, but for plotting lines it's better to leave as NaN or handle carefully)
# We will just plot as is.

# 4. Visualization
plt.figure(figsize=(14, 8))
sns.lineplot(data=retention_summary, dashes=False, linewidth=2.5)

plt.title('Rata-rata Customer Retention Rate (CRR) per Segmen ', fontsize=16)
plt.xlabel('Bulan ke-X (Umur Pelanggan)', fontsize=12)
plt.ylabel('Retention Rate (%)', fontsize=12)
plt.axhline(y=85, color='r', linestyle='--', label='Benchmark (85%)', alpha=0.7)
plt.legend(title='Segmen', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)
plt.ylim(0, 300) # Set limit to see clearly
plt.xlim(1, 48) # Limit to 4 years/48 months as per usual cohort analysis
plt.tight_layout()
plt.savefig('full_retention_curves.png')
plt.show()

```

###NRR	(Net Revenue Retention)
Persentase pendapatan yang dipertahankan dari basis pelanggan yang sudah ada dalam periode waktu tertentu

```python
# --- STEP 6b: FULL NET REVENUE RETENTION (NRR) - 48 MONTHS ---

def plot_full_nrr_heatmap(dataframe, segment_name):
    # Filter Data
    segment_df = dataframe[dataframe['Segment'] == segment_name]

    # Pivot Table: Sum SALES per Cohort Month & Index
    cohort_data = segment_df.groupby(['Cohort Month', 'Cohort Index'])['Sales'].sum().reset_index()
    cohort_revenue = cohort_data.pivot(index='Cohort Month', columns='Cohort Index', values='Sales')

    # Calculate NRR (Divide by Month 1 Revenue)
    initial_revenue = cohort_revenue.iloc[:, 0]
    nrr = cohort_revenue.divide(initial_revenue, axis=0)

    # --- VISUALIZATION SETUP FOR FULL 48 MONTHS ---

    # Calculate dimensions
    n_rows = len(nrr)
    n_cols = nrr.shape[1] # Approx 48

    # Adjust figure size for full view
    fig_width = max(20, n_cols * 0.6)
    fig_height = max(8, n_rows * 0.4 + 2)

    plt.figure(figsize=(fig_width, fig_height))

    plt.title(f' Net Revenue Retention (NRR): {segment_name} (All Months)', fontsize=20, pad=20)

    # Format Y-axis labels
    y_labels = [pd.to_datetime(t).strftime('%Y-%m') for t in nrr.index]

    sns.heatmap(nrr,
                annot=True,
                fmt='.0%',
                cmap='RdYlGn', # Red-Yellow-Green
                vmin=0.0,
                vmax=1.5, # Cap at 150%
                center=1.0, # Yellow is break-even (100%)
                yticklabels=y_labels,
                annot_kws={"size": 8}, # Smaller font for dense grid
                cbar_kws={'label': 'Revenue Retention %'},
                linewidths=0.1,
                linecolor='white'
               )

    plt.ylabel('Cohort Month', fontsize=14)
    plt.xlabel('Cohort Index (Months)', fontsize=14)
    plt.yticks(rotation=0)

    plt.tight_layout()
    plt.savefig(f'nrr_heatmap_{segment_name}_full.png')
    plt.show()

# Execute for all segments
segments = ['SMB', 'Strategic', 'Enterprise']
for seg in segments:
    print(f"Generating NRR Heatmap for: {seg}")
    plot_full_nrr_heatmap(df, seg)
```

```python
df['Cost'] = df['Sales'] - df['Profit']
df['List_Sales'] = df.apply(lambda row: row['Sales'] / (1 - row['Discount']) if row['Discount'] < 1 else 0, axis=1)

# 3. Apply the "Max Discount 20%" Rule
# If discount > 0.2, set it to 0.2. Otherwise keep it.
df['New_Discount'] = df['Discount'].apply(lambda x: 0.20 if x > 0.20 else x)

# 4. Recalculate Sales and Profit based on New Discount
df['New_Sales'] = df['List_Sales'] * (1 - df['New_Discount'])
df['New_Profit'] = df['New_Sales'] - df['Cost']

# 5. Analyze the Impact
comparison = df.groupby('Segment').agg({
    'Profit': 'sum',
    'New_Profit': 'sum',
    'Sales': 'sum',
    'New_Sales': 'sum'
}).reset_index()

comparison['Profit_Diff'] = comparison['New_Profit'] - comparison['Profit']
comparison['Profit_Growth_%'] = (comparison['Profit_Diff'] / comparison['Profit']) * 100

# Total Impact
total_original_profit = df['Profit'].sum()
total_new_profit = df['New_Profit'].sum()
total_growth = ((total_new_profit - total_original_profit) / total_original_profit) * 100

# 6. Visualization
plt.figure(figsize=(12, 6))

# Prepare data for plotting
plot_data = comparison.melt(id_vars='Segment', value_vars=['Profit', 'New_Profit'], var_name='Scenario', value_name='Total Profit')

# Bar Chart
sns.barplot(data=plot_data, x='Segment', y='Total Profit', hue='Scenario', palette=['#E74C3C', '#2ECC71']) # Red for old, Green for new
plt.title(f'Profit Comparison: Actual vs Simulated (Max Discount 20%)\nTotal Profit Growth: {total_growth:.2f}%')
plt.ylabel('Total Profit ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('discount_simulation.png')

# Output Results
print("--- Hasil Simulasi: Diskon Maksimal 20% ---")
print(f"Total Original Profit: ${total_original_profit:,.2f}")
print(f"Total New Profit:      ${total_new_profit:,.2f}")
print(f"Net Increase:          ${total_new_profit - total_original_profit:,.2f} (+{total_growth:.2f}%)")
print("\n--- Detail Per Segmen ---")
print(comparison[['Segment', 'Profit', 'New_Profit', 'Profit_Diff', 'Profit_Growth_%']])

# Check how many transactions were affected
affected_tx = df[df['Discount'] > 0.2].shape[0]
print(f"\nJumlah transaksi yang terkena capping (diskon awal > 20%): {affected_tx} transaksi")
```

```python


# --- 1. Hitung Skenario STOP LOSS (Cut Loss) ---
# Logika: Hanya ambil transaksi yang untung (Profit > 0). Yang rugi dibuang (dianggap 0).
profit_stop_loss = df[df['Profit'] > 0]['Profit'].sum()

# --- 2. Hitung Skenario REPRICING (Max Diskon 20%) ---
# Logika: Diskon > 20% diubah jadi 20%. Hitung ulang profit.
# Cost = Sales - Profit
df['Cost'] = df['Sales'] - df['Profit']
# List Price = Sales / (1 - Discount)
df['List_Sales'] = df.apply(lambda row: row['Sales'] / (1 - row['Discount']) if row['Discount'] < 1 else 0, axis=1)

# Apply Rule
df['New_Discount'] = df['Discount'].apply(lambda x: 0.20 if x > 0.20 else x)
df['New_Sales'] = df['List_Sales'] * (1 - df['New_Discount'])
df['New_Profit'] = df['New_Sales'] - df['Cost']

profit_repricing = df['New_Profit'].sum()

print("--- VERIFIKASI AKHIR ---")
print(f"1. Skenario STOP LOSS (Buang yang rugi): ${profit_stop_loss:,.0f}")
print(f"2. Skenario REPRICING (disc max):   ${profit_repricing:,.0f}")

if profit_repricing > profit_stop_loss:
    print("\nKESIMPULAN: Repricing lebih besar karena mengubah 'Rugi' jadi 'Untung' + menambah margin profit yang sudah ada.")
else:
    print("\nKESIMPULAN: Stop Loss lebih besar.")
```
